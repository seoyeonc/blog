# Fivenodes

```python
data_fivenodes= pd.read_csv('./simulation_results/Real_simulation/fivedones_Simulation.csv')
```

## Baseline

```python
pd.merge(data_fivenodes.query("mtype!='rand' and mtype!='block'").groupby(['nof_filters','method','lags'])['mse'].mean().reset_index(),
         data_fivenodes.query("mtype!='rand' and mtype!='block'").groupby(['nof_filters','method','lags'])['mse'].std().reset_index(),
         on=['method','nof_filters','lags']).rename(columns={'mse_x':'mean','mse_y':'std'}).round(3)
```

```python
pd.merge(data_fivenodes.query("mtype!='rand' and mtype!='block'").groupby(['nof_filters','lags'])['mse'].mean().reset_index(),
         data_fivenodes.query("mtype!='rand' and mtype!='block'").groupby(['nof_filters','lags'])['mse'].std().reset_index(),
         on=['nof_filters','lags']).rename(columns={'mse_x':'mean','mse_y':'std'}).round(3)
```

## Random

```python
data_fivenodes.query("mtype=='rand'and method=='GNAR'")['mse'].unique().round(2)
```

```python
pd.merge(data_fivenodes.query("mtype=='rand'").groupby(['mrate','nof_filters','method','lags'])['mse'].mean().reset_index(),
         data_fivenodes.query("mtype=='rand'").groupby(['mrate','nof_filters','method','lags'])['mse'].std().reset_index(),
         on=['method','nof_filters','mrate','lags']).rename(columns={'mse_x':'mean','mse_y':'std'}).round(3).query("nof_filters==12")
```

```python
pd.merge(data_fivenodes.query("mtype=='rand'").groupby(['mrate','nof_filters','method','lags'])['mse'].mean().reset_index(),
         data_fivenodes.query("mtype=='rand'").groupby(['mrate','nof_filters','method','lags'])['mse'].std().reset_index(),
         on=['method','nof_filters','mrate','lags']).rename(columns={'mse_x':'mean','mse_y':'std'}).round(3).query("nof_filters!=12")
```

## Block

```python
data_fivenodes.query("mtype=='block'and method=='GNAR'")['mse'].unique().round(2)
```

```python
pd.merge(data_fivenodes.query("mtype=='block'").groupby(['mrate','nof_filters','method'])['mse'].mean().reset_index(),
         data_fivenodes.query("mtype=='block'").groupby(['mrate','nof_filters','method'])['mse'].std().reset_index(),
         on=['method','nof_filters','mrate']).rename(columns={'mse_x':'mean','mse_y':'std'}).round(3)
```
