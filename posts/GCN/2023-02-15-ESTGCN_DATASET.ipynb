# Import

```python
import pickle

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
def load_data(fname):
    with open(fname, 'rb') as outfile:
        data_dict = pickle.load(outfile)
    return data_dict
```

```python
def save_data(data_dict,fname):
    with open(fname,'wb') as outfile:
        pickle.dump(data_dict,outfile)
```
